```python
import pandas as pd
import numpy as np
from functools import reduce


diseases = ["measles", "mumps", "rubella", "hepatitis", "pertussis", "rubella", "smallpox", "polio"]
DFList = []

for item in diseases:
    print(item)
```

```python
for item in diseases:
    
    #READ CSV TO DATAFRAME
    #establish csv read path
    inputPath = "dataSource/contagious-diseases/" + item + ".csv"
    #read csv
    data = pd.read_csv(inputPath) 
    #show data head
    data.head()
    #confirm data read
    print("read: " + inputPath)
    
    
    #GET YEAR FOR EACH ENTRY AND GROUPBY FOR SUM OF CASES
    #extract year from year-week date and add to new column
    data["year"] = data['week'].astype(str).str[:-2].astype(np.int64)
    #print new table with year column
    data.head()
    #group by with sum cases per year and state_name
    data_append = pd.DataFrame(data.groupby(["state_name", "year"])["cases"].sum())
    #show data head after groupby
    data_append.head()
    
    
    #APPEND ALL DFs TO A LIST
    #establish DF name
    DFName = item + "DF"
    #select columns from data
    data_forDFList = data [["state", "year", "cases"]]
    #convert to concise dataframe with new column names
    data_forDFList = data_forDFList.rename(columns={'state': 'state', 'year': "year", "cases": item + "_cases"})
    #convert data to DF
    data_forDFList.name = DFName
    #comfirm data converted to DFName
    print(data_forDFList.name + " = data")
    #append data with name to DFList
    DFList.append(data_forDFList)
    
    
    #WRITE CSV TO DATAFRAME
    #establish csv write path
    outputPath = "inputData/" + item + "_byYear_cases.csv"
    #write csv
    data_append.to_csv(outputPath)
    #confirm data write
    print("written: " + outputPath + "\n")   
    
for i in DFList:
    print(i.head())
```

```python
for item in diseases:
    
    #READ CSV TO DATAFRAME
    #establish csv read path
    inputPath = "dataSource/contagious-diseases/" + item + ".csv"
    #read csv
    data = pd.read_csv(inputPath) 
    #show data head
    data.head()
    #confirm data read
    print("read: " + inputPath)
    
    
    #GET YEAR FOR EACH ENTRY AND GROUPBY FOR SUM OF CASES
    #extract year from year-week date and add to new column
    data["year"] = data['week'].astype(str).str[:-2].astype(np.int64)
    #print new table with year column
    data.head()
    #group by with sum cases per year and state_name
    data_append = pd.DataFrame(data.groupby(["state_name", "year"])["incidence_per_capita"].mean())
    #show data head after groupby
    data_append.head()
    
    
    #APPEND ALL DFs TO A LIST
    #establish DF name
    DFName = item + "DF"
    #select columns from data
    data_forDFList = data [["state", "year", "incidence_per_capita"]]
    #convert to concise dataframe with new column names
    data_forDFList = data_forDFList.rename(columns={'state': 'state', 'year': "year", "incidence_per_capita": item + "_casesPerCapita"})
    #convert data to DF
    data_forDFList.name = DFName
    #comfirm data converted to DFName
    print(data_forDFList.name + " = data")
    #append data with name to DFList
    DFList.append(data_forDFList)
    
    
    #WRITE CSV TO DATAFRAME
    #establish csv write path
    outputPath = "inputData/" + item + "_byYear_casesPerCapita.csv"
    #write csv
    data_append.to_csv(outputPath)
    #confirm data write
    print("written: " + outputPath + "\n")   
    
for i in DFList:
    print(i.head())
```
